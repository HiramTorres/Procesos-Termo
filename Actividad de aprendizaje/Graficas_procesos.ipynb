```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(0, 10)
y = 2 / x
P = [2,1]
V = [1,1]


plt.figure(figsize= (6,6))
##plt.plot(x,y, linestyle = "--", color = "c",label = "Proceso 1-2")
plt.plot(V,P, marker = "o",linestyle = "--", color = "c")
plt.xlabel("Volumen")
plt.ylabel("Presión")
plt.xlim(0,3)
plt.ylim(0,3)
plt.yticks([1,2])
plt.xticks([1,2])
plt.legend(loc= "upper right")
plt.annotate("Proceso isocórico", xy=(1.5,1.5),textcoords='offset points')
plt.savefig("Proceso isocórico.jpg")

plt.show()
```
